```python
df = pd.read_csv('../input/xhs_T1_sample5.csv')
```

```python
def extract_words_from_content(content):
    # 使用正则表达式切割字符串
    # 初始化结果字典
    result = defaultdict(lambda: defaultdict(int))

    # 需要排除的字符
    exclude_chars = set('的是了我')

    # 遍历每个段落，提取词语并统计频率
    for segment in content:
        for i in range(len(segment)):
            for j in range(2, 5):
                if i + j <= len(segment):
                    word = segment[i:i+j]
                    first_char = word[0]
                    if first_char not in exclude_chars:
                        result[first_char][word] += 1

    # 将结果转换为普通字典
    return {k: dict(v) for k, v in result.items()}
```

```python

import json
from pandarallel import pandarallel

pandarallel.initialize(progress_bar=True)

# 初始化最终结果字典
final_result = defaultdict(lambda: defaultdict(int))

# 使用 apply 方法遍历 content 列，调用函数生成结果并合并到最终结果字典
results = df[:1000].parallel_apply(lambda x:extract_words_from_content(eval(x['content_list'])),axis=1)

for result in results:
    for first_char, words_freq in result.items():
        for word, freq in words_freq.items():
            final_result[first_char][word] += freq

# 将最终结果转换为普通字典
final_result = {k: dict(v) for k, v in final_result.items()}

with open('../output/words_freq.json', 'w+') as f:
    json.dump(final_result, f,ensure_ascii=False)
# print(final_result)
```
